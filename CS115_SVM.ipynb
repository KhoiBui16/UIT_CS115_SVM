# SVM without Kernels

<font size ='10'>SVM without Kernels</font>

```python
class SVM_without_kernel:

    def __init__(self, learning_rate=0.001, lambda_param=0.01, n_iters=1000):
        self.lr = learning_rate
        self.lambda_param = lambda_param  # Tham số lambda để điều chỉnh độ mạnh của regularization - trade-off
        self.n_iters = n_iters
        self.w = None
        self.b = None

    def fit(self, X, y):
        y_ = np.where(y <= 0, -1, 1)
        n_samples, n_features = X.shape

        # Khởi tạo trọng số và bias
        self.w = np.zeros(n_features)
        self.b = 0

        # Gradient descent
        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):

                # Tính toán điều kiện dual và đạo hàm
                condition = y_[idx] * (np.dot(x_i, self.w) - self.b)  # Điều kiện dual: y_i * (x_i * w - b)

                # Nếu điều kiện thỏa mãn (khoảng cách từ điểm tới siêu phẳng >= 1), sử dụng regularization
                if condition >= 1:
                    dw = 2 * self.lambda_param * self.w
                    db = 0

                else: # Nếu không thỏa mãn điều kiện (chỉ thị điểm nằm trong vùng margin), tính đạo hàm

                    dw = 2 * self.lambda_param * self.w - np.dot(x_i, y_[idx])
                    db = y_[idx]

                # Cập nhật trọng số và bias
                self.w -= self.lr * dw  # Cập nhật trọng số (học theo gradient)
                self.b -= self.lr * db  # Cập nhật bias

    # Dự đoán nhãn cho dữ liệu đầu vào X
    def predict(self, X):
        return np.sign(self._score_funct(X))  # Trả về dấu (1 hoặc -1) của hàm điểm

    # Tính toán giá trị hàm điểm (x * w - b)
    def _score_funct(self, X):
        return np.dot(X, self.w) - self.b  # Đánh giá điểm cho mỗi mẫu dữ liệu

    # Trả về trọng số và bias của mô hình
    def get_weights(self):

        return self.w, self.b  # Trả về các trọng số và bias đã được học
```

# Accuracy:

```python
def accuracy(y_true, y_pred):
  accuracy = np.sum(y_true == y_pred) / len(y_true)
  return accuracy
```

# Plot SVM without Kernels

```python
# Tính giá trị của siêu phẳng (hyperplane) tại một giá trị X (tọa độ x-axis)
def get_hyperplane_val(X, w, b, offset):
    return (-w[0] * X + b + offset) / w[1]
```

```python
def plot_SVM(X, y, w, b, title="Plot SVM"):
    # Khởi tạo một đối tượng figure mới để vẽ đồ thị
    fig = plt.figure()

    # Thêm một subplot vào figure
    ax = fig.add_subplot(1, 1, 1)

    # Tô màu các điểm dữ liệu dựa trên nhãn y, sử dụng colormap 'winter' để phân loại
    plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, cmap='winter', alpha=0.7)

    # Tính toán các giá trị của hyperplane cho các điểm x0 (trục x)
    # x0_1 và x0_2 lần lượt là giá trị nhỏ nhất và lớn nhất của trục x
    x0_1 = np.amin(X[:, 0])
    x0_2 = np.amax(X[:, 0])

    # Tính toán các giá trị tương ứng của trục y cho siêu phẳng chính (decision boundary)
    x1_1 = get_hyperplane_val(x0_1, w, b, 0)
    x1_2 = get_hyperplane_val(x0_2, w, b, 0)

    # Tính toán các giá trị y cho các đường margin âm (offset = -1) và margin dương (offset = 1)
    x1_1_m = get_hyperplane_val(x0_1, w, b, -1)
    x1_2_m = get_hyperplane_val(x0_2, w, b, -1)

    # Tính toán các giá trị y cho các đường margin dương (offset = 1)
    x1_1_p = get_hyperplane_val(x0_1, w, b, 1)
    x1_2_p = get_hyperplane_val(x0_2, w, b, 1)

    # Vẽ các đường phân chia và các đường margin
    # 'y--' là kiểu đường chấm gạch màu vàng cho siêu phẳng chính
    # 'k' là màu đen cho các đường margin
    ax.plot([x0_1, x0_2], [x1_1, x1_2], 'y--')   # Vẽ siêu phẳng chính (decision boundary)
    ax.plot([x0_1, x0_2], [x1_1_m, x1_2_m], 'k')  # Vẽ margin âm
    ax.plot([x0_1, x0_2], [x1_1_p, x1_2_p], 'k')  # Vẽ margin dương

    # Thiết lập phạm vi y-axis sao cho nó đủ rộng để hiển thị các đường margin
    x1_min = np.amin(X[:, 1])
    x1_max = np.amax(X[:, 1])
    ax.set_ylim([x1_min - 3, x1_max + 3])  # Mở rộng phạm vi y-axis để đường margin không bị cắt

    # Thiết lập tiêu đề cho đồ thị
    plt.title(title)
    # Hiển thị đồ thị
    plt.show()
```

# SVM using Kernels

<font size='10'>SVM using Kernels</font>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

class SVM_Dual:
    def __init__(self, kernel='poly', degree=2, gamma='scale', n_iters=1000, learning_rate=0.0001, coef0=1.0):
        # Khởi tạo các tham số của SVM
        self.alpha = None
        self.b = 0
        self.degree = degree                # kernel polynomial
        self.gamma = gamma                  # kernel RBF hoặc scale
        self.n_iters = n_iters
        self.learning_rate = learning_rate
        self.coef0 = coef0                  # kernel polynomial và sigmoid
        self.scaler = StandardScaler()      # Chuẩn hóa dữ liệu

        # Chọn kernel dựa trên tham số 'kernel'
        if kernel == 'poly':
            self.kernel = self.polynomial_kernel  # Dùng kernel polynomial nếu kernel là 'poly'
        elif kernel == 'rbf':
            self.kernel = self.gaussian_kernel  # Dùng kernel Gaussian (RBF) nếu kernel là 'rbf'
        elif kernel == 'sigmoid':
            self.kernel = self.sigmoid_kernel  # Dùng kernel sigmoid nếu kernel là 'sigmoid'

    def polynomial_kernel(self, X, Z):
        # Polynomial kernel: (coef0 + gamma * X.dot(Z.T)) ^ degree
        return (self.coef0 + self.gamma * X.dot(Z.T)) ** self.degree

    def sigmoid_kernel(self, X, Z):
        # Sigmoid kernel: tanh(gamma * X.dot(Z.T) + coef0)
        return np.tanh(self.gamma * X.dot(Z.T) + self.coef0)

    def gaussian_kernel(self, X, Z):
        # Gaussian (RBF) kernel: exp(-gamma * ||X - Z||^2)
        return np.exp(-self.gamma * np.linalg.norm(X[:, np.newaxis] - Z[np.newaxis, :], axis=2) ** 2)

    def fit(self, X, y):
        # Chuẩn hóa dữ liệu
        X = self.scaler.fit_transform(X)
        self.X = X
        self.y = y

        # Khởi tạo giá trị alpha và bias
        self.alpha = np.random.random(X.shape[0])  # Ngẫu nhiên hóa giá trị alpha
        self.b = 0
        self.ones = np.ones(X.shape[0])  # Mảng chứa giá trị 1

        # Tính gamma nếu gamma là 'scale'
        if self.gamma == 'scale':
            self.gamma = 1 / (X.shape[1] * np.var(X))
        elif self.gamma == 'auto':
            self.gamma = 1 / X.shape[1]

        # Tạo ma trận kernel K(xi, xj) với yếu tố yi * yj
        kernel_matrix = np.outer(y, y) * self.kernel(X, X)  # yi * yj * K(xi, xj)

        for i in range(self.n_iters):
            # Tính gradient của hàm mục tiêu (objective function)
            gradient = self.ones - kernel_matrix.dot(self.alpha) # 1 – yk ∑ αj yj K(xj, xk)
            self.alpha += self.learning_rate * gradient  # α = α + η*(1 – yk ∑ αj yj K(xj, xk)) để tối ưu cực đại

            # Giới hạn alpha trong khoảng (0, coef0)
            self.alpha[self.alpha > self.coef0] = self.coef0
            self.alpha[self.alpha < 0] = 0

            # Tính loss (optional), đây là hàm mục tiêu cần tối đa hóa
            # ∑αi – (1/2) ∑i ∑j αi αj yi yj K(xi, xj)
            loss = np.sum(self.alpha) - 0.5 * np.sum(np.outer(self.alpha, self.alpha) * kernel_matrix)

        # Lọc chỉ giữ các alpha có giá trị trong khoảng (0, coef0)
        valid_alpha_indices = np.where((self.alpha > 0) & (self.alpha < self.coef0))[0]

        # Tính b (bias) từ các alpha không phải biên (0 < α < coef0)
        b_list = []
        for index in valid_alpha_indices:
            b_list.append(y[index] - (self.alpha * y).dot(self.kernel(X, X[index])))

        # Nếu b_list không rỗng, tính giá trị b bằng trung bình của b_list
        if len(b_list) > 0:
            self.b = np.mean(b_list)
        else:
            self.b = 0  # Nếu không có giá trị b hợp lệ, gán b = 0

    def predict(self, X):
        # Chuẩn hóa dữ liệu trước khi dự đoán
        X = self.scaler.transform(X)

        return np.sign(self.decision_function(X))  # Trả về dấu của giá trị quyết định

    def score(self, X, y):
        y_hat = self.predict(X)  # Dự đoán nhãn
        return np.mean(y == y_hat)  # Tính tỷ lệ dự đoán đúng

    def decision_function(self, X):
        return (self.alpha * self.y).dot(self.kernel(self.X, X)) + self.b  # Tính giá trị quyết định f(x)

    def plot(self, title='SVM Plot'):
        self.X = self.scaler.fit_transform(self.X)  # Chuẩn hóa dữ liệu
        plt.scatter(self.X[:, 0], self.X[:, 1], c=self.y, s=50, cmap='winter', alpha=.8)

        # Lấy đối tượng trục của đồ thị
        ax = plt.gca()

        # Lấy phạm vi x-axis
        xlim = ax.get_xlim()

        # Lấy phạm vi y-axis
        ylim = ax.get_ylim()

        # Tạo lưới để tính giá trị quyết định cho từng điểm trong lưới
        xx = np.linspace(xlim[0], xlim[1], 50)  # Tạo các điểm trên trục x
        yy = np.linspace(ylim[0], ylim[1], 50)  # Tạo các điểm trên trục y

        # Tạo lưới 2D từ các điểm trên trục x và y
        YY, XX = np.meshgrid(yy, xx)

        # Chuyển đổi lưới thành dạng mảng
        xy = np.vstack([XX.ravel(), YY.ravel()]).T

        # Tính giá trị quyết định cho mỗi điểm trong lưới
        Z = self.decision_function(xy).reshape(XX.shape)

        # Vẽ các đường contour cho các giá trị quyết định (ở các mức -1, 0, 1)
        ax.contour(XX, YY, Z, levels=[-1, 0, 1], linestyles=['--', '-', '--'])

        # Thêm tiêu đề cho đồ thị
        plt.title(title)
        # Hiển thị đồ thị
        plt.show()
```

# SVM in Scikit-Learn

```python
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt
```

## Vẽ đồ thị

```python
def plot(X, y, svm, title='SVM plot'):
    # Vẽ scatter plot cho các điểm dữ liệu X, y
    plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.winter, alpha=0.7)


    # Lấy trục hiện tại (axis) để điều chỉnh phạm vi
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # Tạo lưới để tính toán giá trị quyết định (decision function) của mô hình
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)

    # Tạo lưới 2D từ các điểm của trục x và y
    YY, XX = np.meshgrid(yy, xx)

    # Chuyển đổi lưới thành mảng 2 chiều, mỗi hàng là một điểm trong lưới
    xy = np.vstack([XX.ravel(), YY.ravel()]).T

    # Tính giá trị quyết định cho từng điểm trong lưới
    # Sử dụng phương thức decision_function từ mô hình SVM
    Z = svm.decision_function(xy).reshape(XX.shape)

    # Vẽ các đường phân chia (decision boundary) và các đường margin (các đường mức -1 và +1)
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])

    # Vẽ các support vector, là những điểm hỗ trợ trong quá trình huấn luyện của mô hình SVM
    ax.scatter(svm.support_vectors_[:, 0], svm.support_vectors_[:, 1], s=100, linewidth=1, facecolors='none', edgecolors='k')

    # Thêm tiêu đề cho đồ thị
    plt.title(title)
    # Hiển thị đồ thị
    plt.show()
```

## Bước 2: Xây dựng BinarySVM và MultiClassSVM

### 2.1) BinarySVM

```python
class BinarySVM:
    def __init__(self, gamma=1.0, coef0=1.0, n_iters=1000, learning_rate=0.01, kernel='linear', degree=3):
        self.gamma = gamma                  # kernel RBF hoặc Poly
        self.coef0 = coef0                  # Offset cho kernel Poly hoặc Sigmoid
        self.n_iters = n_iters
        self.learning_rate = learning_rate
        self.kernel_type = kernel           # Loại kernel sử dụng (linear, poly, rbf)
        self.degree = degree                # Bậc của kernel đa thức (poly)
        self.alpha = None
        self.b = 0

    def _linear_kernel(self, X, Z):
        # Kernel tuyến tính: K(X, Z) = X * Z.T
        return np.dot(X, Z.T)

    def _polynomial_kernel(self, X, Z):
        # Kernel đa thức: K(X, Z) = (coef0 + gamma * X * Z.T)^degree
        return (self.coef0 + self.gamma * np.dot(X, Z.T)) ** self.degree

    def _rbf_kernel(self, X, Z):
        # Kernel RBF (Gaussian): K(X, Z) = exp(-gamma * ||X - Z||^2)
        return np.exp(-self.gamma * np.linalg.norm(X[:, np.newaxis] - Z[np.newaxis, :], axis=2) ** 2)

    def _select_kernel(self):
        if self.kernel_type == 'linear':
            return self._linear_kernel
        elif self.kernel_type == 'poly':
            return self._polynomial_kernel
        elif self.kernel_type == 'rbf':
            return self._rbf_kernel
        else:
            raise ValueError("Unsupported kernel type")  # Báo lỗi nếu kernel không hợp lệ

    def fit(self, X, y):
        # Huấn luyện SVM nhị phân với dữ liệu X và nhãn y
        self.X = X  # Lưu dữ liệu huấn luyện
        self.y = y  # Lưu nhãn nhị phân
        self.alpha = np.random.random(X.shape[0])
        self.b = 0

        # Lấy hàm kernel dựa trên lựa chọn
        kernel = self._select_kernel()

        # Ma trận kernel có tính đến nhãn y
        y_mul_kernel = np.outer(y, y) * kernel(X, X)

        for _ in range(self.n_iters):
            # Tính gradient
            gradient = 1 - y_mul_kernel.dot(self.alpha)

            # Cập nhật alpha
            self.alpha += self.learning_rate * gradient

            # Đảm bảo alpha không âm
            self.alpha[self.alpha < 0] = 0

        # Xác định các điểm hỗ trợ (support vectors)
        support_indices = np.where((self.alpha > 0) & (self.alpha < 1))[0]

        # Tính bias dựa trên các điểm hỗ trợ
        self.b = np.mean(y[support_indices] - (self.alpha * y).dot(kernel(X, X[support_indices])))

    def decision_function(self, X):
        kernel = self._select_kernel()  # Lấy hàm kernel đã chọn
        return (self.alpha * self.y).dot(kernel(self.X, X)) + self.b  # Tính điểm số quyết định
```

### 2.2) Multi-class SVM

```python
class MultiClassSVM:
    # Khởi tạo Multi-Class SVM với chiến lược One-vs-All
    def __init__(self, gamma=1.0, coef0=1.0, n_iters=1000, learning_rate=0.01, kernel='linear', degree=3):

        self.gamma = gamma                  # kernel RBF hoặc Poly
        self.coef0 = coef0                  # Offset cho kernel Poly hoặc Sigmoid
        self.n_iters = n_iters
        self.learning_rate = learning_rate
        self.kernel_type = kernel           # kernel sử dụng (linear, poly, rbf)
        self.degree = degree                # Bậc của kernel Poly
        self.classifiers = []               # Danh sách lưu các mô hình nhị phân

    def fit(self, X, y):
        # Lấy danh sách các lớp trong dữ liệu
        self.classes = np.unique(y)

        # Khởi tạo danh sách các mô hình nhị phân
        self.classifiers = []

        # Với mỗi lớp
        for cls in self.classes:
            # Chuyển nhãn thành nhị phân (1/-1)
            binary_y = np.where(y == cls, 1, -1)

            # Tạo một mô hình SVM nhị phân
            binary_svm = BinarySVM(
                gamma=self.gamma, coef0=self.coef0, n_iters=self.n_iters,
                learning_rate=self.learning_rate, kernel=self.kernel_type, degree=self.degree
            )

            # Huấn luyện SVM nhị phân
            binary_svm.fit(X, binary_y)

            # Lưu mô hình vào danh sách
            self.classifiers.append(binary_svm)

     # Dự đoán nhãn đa lớp
    def predict(self, X):
        # Tính điểm số
        decision_scores = np.array([clf.decision_function(X) for clf in self.classifiers]).T

        # Trả về nhãn có điểm số cao nhất
        return self.classes[np.argmax(decision_scores, axis=1)]

    # Vẽ vùng quyết định (decision boundaries) cho Multi-class SVM
    def plot(self, X, y):

        # Vẽ dữ liệu gốc
        plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolors='k')
        # Lấy phạm vi trục X
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        # Lấy phạm vi trục Y
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

        # Tạo lưới dự đoán
        xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

        # Dự đoán nhãn cho lưới
        Z = self.predict(np.c_[xx.ravel(), yy.ravel()])

        # Định dạng lại kích thước của Z
        Z = Z.reshape(xx.shape)

        # Vẽ contour vùng quyết định
        plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')

        plt.title("Multi-Class SVM Decision Boundaries")
        plt.show()
```

## Bước 3: Huấn luyện mô hình

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23520761)
svm = MultiClassSVM(kernel='rbf', gamma=0.5, coef0=1.0, n_iters=1000, learning_rate=0.001)
svm.fit(X_train, y_train)
```

## 3) Xây dựng class Binary SVM và Mulit-SVM

### 3.1) BinarySVM

```python
class BinarySVM:
    def __init__(self, gamma=1.0, coef0=1.0, n_iters=500, learning_rate=0.01, kernel='linear', degree=3):
        self.gamma = gamma                  # kernel RBF hoặc Poly
        self.coef0 = coef0                  # Offset cho kernel Poly hoặc Sigmoid
        self.n_iters = n_iters
        self.learning_rate = learning_rate
        self.kernel_type = kernel           # Loại kernel sử dụng (linear, poly, rbf)
        self.degree = degree                # Bậc của kernel đa thức (poly)
        self.alpha = None
        self.b = 0

    def _linear_kernel(self, X, Z):
        # Kernel tuyến tính: K(X, Z) = X * Z.T
        return np.dot(X, Z.T)

    def _polynomial_kernel(self, X, Z):
        # Kernel đa thức: K(X, Z) = (coef0 + gamma * X * Z.T)^degree
        return (self.coef0 + self.gamma * np.dot(X, Z.T)) ** self.degree

    def _rbf_kernel(self, X, Z):
        # Kernel RBF (Gaussian): K(X, Z) = exp(-gamma * ||X - Z||^2)
        return np.exp(-self.gamma * np.linalg.norm(X[:, np.newaxis] - Z[np.newaxis, :], axis=2) ** 2)

    def _select_kernel(self):
        if self.kernel_type == 'linear':
            return self._linear_kernel
        elif self.kernel_type == 'poly':
            return self._polynomial_kernel
        elif self.kernel_type == 'rbf':
            return self._rbf_kernel
        else:
            raise ValueError("Unsupported kernel type")  # Báo lỗi nếu kernel không hợp lệ

    def fit(self, X, y):
        # Huấn luyện SVM nhị phân với dữ liệu X và nhãn y
        self.X = X  # Lưu dữ liệu huấn luyện
        self.y = y  # Lưu nhãn nhị phân
        self.alpha = np.random.random(X.shape[0])
        self.b = 0

        # Lấy hàm kernel dựa trên lựa chọn
        kernel = self._select_kernel()

        # Ma trận kernel có tính đến nhãn y
        y_mul_kernel = np.outer(y, y) * kernel(X, X)

        for _ in range(self.n_iters):
            # Tính gradient
            gradient = 1 - y_mul_kernel.dot(self.alpha)
            gradient = np.clip(gradient, -1e5, 1e5)

            # Cập nhật alpha
            self.alpha += self.learning_rate * gradient

            # Đảm bảo alpha không âm
            self.alpha[self.alpha < 0] = 0

        # Xác định các điểm hỗ trợ (support vectors)
        support_indices = np.where((self.alpha > 0) & (self.alpha < 1))[0]

        # Tính bias dựa trên các điểm hỗ trợ
        if len(support_indices) == 0:
            self.b = 0
        else:
            self.b = np.mean(y[support_indices] - (self.alpha * y).dot(kernel(X, X[support_indices])))


    def decision_function(self, X):
        kernel = self._select_kernel()  # Lấy hàm kernel đã chọn
        return (self.alpha * self.y).dot(kernel(self.X, X)) + self.b  # Tính điểm số quyết định
```

### 3.2) Multi-SVM

```python
class MultiClassSVM:
    # Khởi tạo Multi-Class SVM với chiến lược One-vs-All
    def __init__(self, gamma=1.0, coef0=1.0, n_iters=500, learning_rate=0.001, kernel='linear', degree=3):

        self.gamma = gamma                  # kernel RBF hoặc Poly
        self.coef0 = coef0                  # Offset cho kernel Poly hoặc Sigmoid
        self.n_iters = n_iters
        self.learning_rate = learning_rate
        self.kernel_type = kernel           # kernel sử dụng (linear, poly, rbf)
        self.degree = degree                # Bậc của kernel Poly
        self.classifiers = []               # Danh sách lưu các mô hình nhị phân

    def fit(self, X, y):
        # Lấy danh sách các lớp trong dữ liệu
        self.classes = np.unique(y)

        # Khởi tạo danh sách các mô hình nhị phân
        self.classifiers = []

        # Với mỗi lớp
        for cls in self.classes:
            # Chuyển nhãn thành nhị phân (1/-1)
            binary_y = np.where(y == cls, 1, -1)

            # Tạo một mô hình SVM nhị phân
            binary_svm = BinarySVM(
                gamma=self.gamma, coef0=self.coef0, n_iters=self.n_iters,
                learning_rate=self.learning_rate, kernel=self.kernel_type, degree=self.degree
            )

            # Huấn luyện SVM nhị phân
            binary_svm.fit(X, binary_y)

            # Lưu mô hình vào danh sách
            self.classifiers.append(binary_svm)

     # Dự đoán nhãn đa lớp
    def predict(self, X):
        # Tính điểm số
        decision_scores = np.array([clf.decision_function(X) for clf in self.classifiers]).T

        # Trả về nhãn có điểm số cao nhất
        return self.classes[np.argmax(decision_scores, axis=1)]

    # Vẽ vùng quyết định (decision boundaries) cho Multi-class SVM
    def plot(self, X, y):

        # Vẽ dữ liệu gốc
        plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolors='k')
        # Lấy phạm vi trục X
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        # Lấy phạm vi trục Y
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

        # Tạo lưới dự đoán
        xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

        # Dự đoán nhãn cho lưới
        Z = self.predict(np.c_[xx.ravel(), yy.ravel()])

        # Định dạng lại kích thước của Z
        Z = Z.reshape(xx.shape)

        # Vẽ contour vùng quyết định
        plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')

        plt.title("Multi-Class SVM Decision Boundaries")
        plt.show()
```

# Vì sao SVM tự viết có độ chính xác cao hơn scikit-learn?


## 1. **Ảnh hưởng của StandardScaler**
   - **Mô hình tự viết**: Luôn sử dụng dữ liệu đã chuẩn hóa (scaled) với StandardScaler, đảm bảo các đặc trưng được đưa về cùng một khoảng giá trị (trung bình 0, độ lệch chuẩn 1). Điều này giúp thuật toán tối ưu hội tụ tốt hơn.
   - **Scikit-learn**: Nếu quên hoặc không áp dụng scaler đúng cách, các đặc trưng có thể có khoảng giá trị khác nhau, dẫn đến việc SVM khó tối ưu hơn và giảm hiệu suất.

## 2. **Tối ưu hóa quá mức (Overfitting)**
   - **Mô hình tự viết**: Thường được thiết kế đơn giản, tập trung tối ưu hóa mạnh mẽ trên tập dữ liệu huấn luyện nhỏ, đôi khi dẫn đến overfitting (khớp quá mức). Kết quả là độ chính xác trên tập kiểm tra có thể cao hơn nhưng dễ mất tổng quát hóa trên tập lớn.
   - **Scikit-learn**: Sử dụng thuật toán tối ưu hóa tổng quát (SMO) để tránh overfitting. Tuy nhiên, trên tập dữ liệu nhỏ, việc tránh overfitting có thể khiến mô hình không đạt hiệu suất tốt nhất trên tập kiểm tra.

## 3. **Điều chỉnh siêu tham số**
   - **Mô hình tự viết**: Các siêu tham số như `learning_rate`, `gamma`, số vòng lặp (`n_iters`) được tinh chỉnh thủ công, phù hợp với tập dữ liệu cụ thể. Điều này tạo lợi thế cho việc tối ưu trên một tập dữ liệu cụ thể.
   - **Scikit-learn**: Yêu cầu người dùng tinh chỉnh thủ công các tham số như `C`, `gamma`, và kernel. Nếu không điều chỉnh cẩn thận, hiệu suất có thể thấp hơn.

## 4. **Độ phức tạp trong thuật toán**
   - **Mô hình tự viết**: Đơn giản hóa thuật toán và không áp dụng nhiều cơ chế phức tạp để đảm bảo tính tổng quát. Điều này giúp mô hình dễ dàng "khớp" với dữ liệu hiện tại, đặc biệt khi dữ liệu ít.
   - **Scikit-learn**: Sử dụng cơ chế điều chỉnh tổng quát hóa, chẳng hạn chuẩn hóa lỗi và soft-margin, giúp mô hình hoạt động ổn định trên dữ liệu đa dạng, nhưng có thể không đạt độ chính xác cao trên tập nhỏ.

## 5. **Dữ liệu sử dụng**
   - Với tập dữ liệu nhỏ như MNIST được lấy mẫu (900 train, 450 test), việc sử dụng một thuật toán tổng quát hóa cao như scikit-learn có thể không tận dụng hết khả năng của mô hình, trong khi thuật toán tự viết dễ dàng tối ưu hóa cục bộ hơn.
